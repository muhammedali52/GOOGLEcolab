```python
# Drive'da bir dosya yolu
dosya_yolu = '/content/drive/MyDrive/muhammedData/ProductPriceIndex.csv'


```

```python
import pandas as pd

# CSV dosyasını okuma
veri = pd.read_csv(dosya_yolu)
```

```python
# Eksik verileri kontrol etme
print("Eksik Veri Sayısı:")
print(veri.isnull().sum())

# Eksik verileri ortalama değerlerle tamamlama

veri_tamamlanan = veri.fillna(veri.mean())

# Tamamlanan veriyi kontrol etme
print("\nEksik Veri Tamamlanan Hali:")
print(veri_tamamlanan.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tarih sütununu datetime tipine çevirme (eğer çevirmeniz gerekiyorsa)
veri_tamamlanan['date'] = pd.to_datetime(veri_tamamlanan['date'])

# Her tarih için alışveriş sayısını hesaplama
alisveris_sayisi = veri_tamamlanan['date'].value_counts()

# En çok alışveriş yapılan 3 tarihi bulma
en_cok_alisveris_yapilan_tarihler = alisveris_sayisi.nlargest(3).index

# Veriyi görselleştirme
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x='date', data=veri_tamamlanan[veri_tamamlanan['date'].isin(en_cok_alisveris_yapilan_tarihler)])
plt.title('En Çok Alışveriş Yapılan Tarihler')
plt.xticks(rotation=45)
plt.show()
```

```python
print(veri_tamamlanan['farmprice'].dtype)
```

```python
# 'farmprice' sütununu string verilere dönüştürme
veri_tamamlanan['farmprice'] = veri_tamamlanan['farmprice'].astype(str)

# 'farmprice' sütunundaki her değerin ilk harfini silme
veri_tamamlanan['farmprice'] = veri_tamamlanan['farmprice'].str[1:]

# 'farmprice' sütununu sayısal verilere dönüştürme
veri_tamamlanan['farmprice'] = pd.to_numeric(veri_tamamlanan['farmprice'], errors='coerce')
```

```python
#farmprice deki en pahallı 5 ürün
# Fiyat sütununu kontrol etme
if 'farmprice' in veri_tamamlanan.columns:
    # En pahalı 5 ürünü bulma
    en_pahali_urunler = veri_tamamlanan.nlargest(5, 'farmprice')

    # Sonuçları gösterme
    print("En Pahalı 5 Ürün:")
    print(en_pahali_urunler[['productname', 'farmprice']])
```

```python
import matplotlib.pyplot as plt

# 'UrunAdi' sütunundaki en fazla adı geçen 5 ürünü bulma
en_fazla_ad_gecen_urunler = veri_tamamlanan['productname'].value_counts().nlargest(5)

# Dairesel grafik oluşturma
plt.figure(figsize=(8, 8))
plt.pie(en_fazla_ad_gecen_urunler, labels=en_fazla_ad_gecen_urunler.index, autopct='%1.1f%%', startangle=140)
plt.title('En Fazla Adı Geçen İlk 5 Ürün')
plt.show()
```
